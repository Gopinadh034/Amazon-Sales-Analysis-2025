```python
#8. how much Revenue lost due to cancelled orders/Pending/Completed
```

```python
cancelled_amount=df[df["Status"]=='Cancelled']["Total Sales"].sum()
print("Cancelled_amount=",cancelled_amount)
pending_amount=df[df["Status"]=='Pending']["Total Sales"].sum()
print("Pending_amount=",pending_amount)
completed_amount=df[df["Status"]=='Completed']["Total Sales"].sum()
print("completed_amount=",completed_amount)
```
